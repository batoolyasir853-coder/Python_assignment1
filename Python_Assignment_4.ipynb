<u>**Introduction:**</u>  
In Python programming, the while loop is used to execute a block of code repeatedly as long as a given condition remains True. When we use while True, it creates an infinite loop that continues until it encounters a break statement.  
Inside these loops, we often use Conditional Statements (if, elif, and else) to control the flow of the program:  
**If/Elif:** These allow the program to check specific conditions and execute different blocks of code based on which condition is met first.  
**Execution Flow:** Python only executes the first condition that evaluates to True and skips the rest of the blocks in that set.  
**The Else Block:** This serves as a final "catch-all" that runs only if none of the preceding conditions are satisfied.





````python
# Teacher's Provided Code
string = "aabbcc"
count = 3

while True:
    if string[0] == 'a':
        string = string[2:]  # Removes first two characters
    elif string[-1] == 'b':
        string = string[:2]  # Keep only first two characters
    else:
        count += 1
        break  # Loop ends here

print("Final String:", string)
print("Final Count:", count)
````

**Description:**  
This program demonstrates the implementation of a continuous while loop designed for Dynamic Data Manipulation. It specifically showcases how to evaluate and modify a string through Slicing and Conditional Branching until a termination condition is triggered.  
<u> **1. Variable Initialization** </u>  
* **Target String:** A string variable is defined with the initial sequence "aabbcc".  
* **Iteration Counter:** The variable count is assigned a starting value of 3, representing a baseline for the final output.  
<u> **2. Detailed Logical Steps** </u>
The program executes within an Indefinite Loop (while True), meaning it will persist in its operations until the logic explicitly calls for a shutdown.  
* **Primary Execution (Iteration 1):**  
The system evaluates the starting index (string[0]). Since the string is "aabbcc", the if condition for 'a' is met.  
* **Action:**  
 The program performs a Front-end Slice (string[2:]), effectively stripping the first two characters.  
* **State Change:**  
 The string is updated to "bbcc".
* **Secondary Execution (Iteration 2):**  
* The loop re-evaluates the modified string "bbcc". The first character is now 'b', making the if statement False.  
* It then checks the elif condition at the final index (string[-1]). The character is 'c', so this condition also fails.  
* **Termination Action:**  
 Since no pattern-match is found, the Else Block is activated.  
* The count is incremented by 1 (reaching 4), and the break command is utilized to cease all further iterations.  
<u> **3. Final Program Output** </u>  
Upon exiting the loop, the system prints the final states of the variables:
* Final String State: bbcc
* Final Counter Value: 4
<u> **4. Technical Concepts Implemented** </u>  
* **Negative Indexing:**  
 Leveraging string[-1] to verify data from the end of the sequence.
* **Infinite Loop Control:**  
 Using while True to handle tasks where the number of cycles isn't predetermined.
* **Data Slicing:**  
 Dynamically resizing a string based on specific logic.

### **Independent Implementation & Original Logic**  
To demonstrate my understanding of while loops, I have designed a "Score Increment System." This example illustrates how a loop monitors a numerical value and terminates precisely when a specific limit is exceeded.

````python
# Simple Score Tracking System
current_score = 0
termination_count = 0

while True:
    if current_score < 10:
        current_score += 5  # Incrementing score by 5
    elif current_score == 10:
        current_score += 2  # Adding a bonus when score hits 10
    else:
        termination_count += 1
        break  # Exit the loop when score exceeds the limit

print("Final Calculated Score:", current_score)
print("Total Termination Attempts:", termination_count)
````

**Description:**  
This code utilizes a while True loop to manage a scoring system. The program evaluates the current state of the data in each cycle and decides whether to increment the value or terminate the process based on predefined thresholds.  
<u> **1. Variable Setup** </u>    
* **current_score:**  
Initially set to 0. It acts as the primary data being processed.  
* **termination_count:**  
 A counter used to track the final execution step of the loop.
<u> **2. Logical Progression** </u>  
* **Iteration 1:**  
 The current_score is 0. Since 0 < 10, the if condition is satisfied, and the score increases to 5.
* **Iteration 2:**  
 The score is now 5. The if condition remains True, and the score increases to 10.
* **Iteration 3:**  
 The score is exactly 10. The if condition fails (10 is not less than 10), so the elif condition is checked. Since score == 10 is True, the score increases to 12.
* **Iteration 4:**  
 The score is now 12. Both if and elif evaluate to False. Consequently, the else block is executed, the termination_count becomes 1, and the break statement terminates the loop.
<u> **3. Execution Results** </u>  
* **Final Score State:** 12  
* **Termination Record:** 1  
<u> **4. Key Technical Takeaway** </u>  
This implementation proves that if-elif-else structures ensure only one logical path is taken per iteration. By integrating a break statement within the else block, we can prevent infinite loops and ensure the program exits safely once the data reaches its final desired state.

### **Break Statement**  
**Definition:**  
The break statement in Python is used to terminate the execution of the current loop immediately. Once the program encounters a break, it jumps out of the loop and starts executing the next line of code written outside the loop.  
**Why we used it in our assignment:**  
In our while True loop, there was no condition at the top to stop the iteration. Therefore, we used break inside the else block as an Exit Strategy.
* Without break, the loop would have become an "Infinite Loop," causing the program to run forever.  
* In the first code, it stopped the loop once the string no longer matched the 'a' or 'b' patterns.  
* In second code, it stopped the loop once the score exceeded our target limit.   

**Syntex**

```
while condition:  
    if specific_condition:    
        break  # Exits the loop immediately



````python
# Program to stop when a specific number is found
for number in range(1, 10):
    if number == 5:
        break  # Stops the loop when number hits 5
    print("Current Number:", number)

print("Loop terminated because break was triggered.")
````

**Description:**  
In this example, the loop is supposed to run from 1 to 9. However, we added an if condition to check if the number is 5. As soon as the condition becomes True, the break statement executes, and the loop stops immediately. As a result, numbers from 6 to 9 are never printed.

### **continue Statement**
**Definition:**  
The continue statement is used to skip the remaining code inside the current iteration and immediately jump back to the top of the loop for the next cycle. Unlike break, it does not terminate the loop; it only bypasses specific steps for a particular turn.  
**Syntax**  
```
while condition:
    if skip_condition:
        continue
    # Code below this will be skipped if continue is triggered

````python
# Program to skip a specific value
for count in range(1, 6):
    if count == 3:
        continue  # Skips the rest of the code when count is 3
    print("Processing Number:", count)
````

**Description:**  
In this example, the loop is designed to iterate from 1 to 5. However, when the variable count reaches 3, the continue statement is triggered. This causes the program to skip the print function for that specific iteration and move directly to the next number (4). As a result, the output shows 1, 2, 4, and 5, while 3 is excluded.
