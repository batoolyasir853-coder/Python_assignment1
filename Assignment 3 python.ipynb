### **<u> 1. Match statement:** </u>      
The Match statement was introduced in python 3.10 as a powerful tool for structured pattern Matching. It allows a variable to be compared against a series of patterens, providing a more readable alternative to complex if-elif-else chains.  
**Key Points:**  
  *   **Structural Pattern Matching:**  
 It allows us to match not just values, but also the structure of the data.
* **Simplified Logic:**  
 It is a cleaner and more readable alternative to using multiple if-elif-else statements.
*   **Case Patterns:**  
 Each possible value is defined using the case keyword.  
*   **Wildcard Operator (_):**  
 The underscore is used as a default case that executes if no other pattern matches.
* **Efficient Execution:**  
 It is generally faster and more organized when dealing with a large number of conditions.

 ### <u> **Syntax:** </u>
 The structure of a match statement is:

* **Match:** This is the main keyword where we put the variable we want to check.
* **Case:** We write `case` followed by the value we are looking for.
* **Colon (`:`):** Every match and case line must end with a colon.
* **Indentation:** The code under each case must be moved to the right (4 spaces).
* **Wildcard (`_`):** This is the "default" case. If no other case matches, this one will run.
*













```
# Simple Syntax Example
color = "red"

match color:
    case "red":
        print("Stop!")
    case "green":
        print("Go!")
    case _:
        print("Invalid Color")
```

**Example 01:  
 Grade Checker:**  
This program takes a grade (A, B, or C) and prints a simple message. It shows how `match` handles specific characters.

```
grade = input("Enter your Grade (A, B, C): ").upper()

match grade:
    case "A":
        print("Excellent work!")
    case "B":
        print("Good job!")
    case "C":
        print("You passed.")
    case _:
        print("Grade not recognized.")
```

**Example 02:  
Simple Calculator Commands:**  
This program helps a user perform a basic task by matching the command word they type.

```
command = input("Type a command (start, stop, exit): ").lower()

match command:
    case "start":
        print("System is starting...")
    case "stop":
        print("System is shutting down.")
    case "exit":
        print("Exiting the program.")
    case _:
        print("I don't understand that command.")
```

**Example 03:  
 Fruit Price Checker:**  
In this program, we use the `match` statement to find the price of a fruit. It shows how we can use the pipe (`|`) operator to give the same price to multiple fruits.

```
# Simple Fruit Price Program
fruit = input("Enter fruit name (Apple, Banana, Mango): ").capitalize()

match fruit:
    case "Apple":
        print("Price: $2 per kg")
    case "Banana" | "Mango":
        print("Price: $1.5 per kg")
    case _:
        print("Sorry, this fruit is not available.")
```

### <u> **2.Comparison: Match vs. If-Else Statement: </u>**
While both structures are used for decision-making, they have distinct characteristics that make them suitable for different scenarios.


| Feature | If-Else Statement | Match Case Statement |
| :--- | :--- | :--- |
| **Logic Type** | It is based on **Boolean logic** (True/False conditions). | It is based on **Structural Pattern Matching**. |
| **Flexibility** | Excellent for **range-based** conditions (e.g., `x > 10`). | Best for **specific values** and matching data structures. |
| **Readability** | Can become messy with many `elif` blocks. | Very **clean and declarative**, easy to maintain. |
| **Performance** | Checks conditions **sequentially** (one by one). | Highly **optimized**; jumps directly to the matching case. |
| **Default Case** | Uses the `else` block as a fallback. | Uses the **wildcard `_`** pattern as a fallback. |
| **Multiple Values**| Requires logical `or` (e.g., `if x == 1 or x == 2`). | Uses the **pipe `|` operator** (e.g., `case 1 | 2:`) for grouping. |
| **Complexity** | Better for **dynamic expressions** and variables. | Better for **fixed patterns** and constant values. |   

### **Comparison in Action:**
To better understand the difference, let's look at the same logic implemented using both the traditional `if-elif-else` structure and the modern `match-case` statement.



```
# --- 1. Using If-Else Statement ---
light = "Red"

print("Checking with If-Else:")
if light == "Red":
    print("Action: Stop")
elif light == "Yellow":
    print("Action: Wait")
elif light == "Green":
    print("Action: Go")
else:
    print("Action: Invalid Light")

print("-" * 20)

# --- 2. Using Match Case Statement ---
print("Checking with Match Case:")
match light:
    case "Red":
        print("Action: Stop")
    case "Yellow":
        print("Action: Wait")
    case "Green":
        print("Action: Go")
    case _:
        print("Action: Invalid Light")
```

### **<u>3. Nested If-Else Statement:</u>**
A Nested If-Else statement is a programming pattern where one conditional block is embedded inside another. It is used to manage multi-layered logic where the second decision is only relevant if the first condition is successfully met.  
**Key Points:**  
*   It enables a hierarchical flow of logic, meaning one check is dependent on the result of a previous one.
* While there is no technical limit to the number of layers, keeping nesting minimal ensures the code remains clean.

* The internal "if" statement is triggered only when the external "if" condition is True.   
*   Strict indentation is mandatory to distinguish which block of code belongs to which specific condition.
*  This structure is perfect for complex verifications like multi-step security or filtered searches.  
**Syntax Structure:**  
**The Outer If:** Begins with the primary if keyword and a condition followed by a colon. It acts as the initial filter.  
**The Indented Inner If:** A secondary if statement placed inside the first block, shifted to the right to show its dependency.  
**The Inner Else:** This handles the scenario where the first gate is passed, but the second specific condition fails.  
**The Outer Else:** Aligned with the very first if, this block executes if the initial condition is False, bypassing everything inside.   
**Indented Code:** All executable actions must be indented under their respective headers to prevent syntax errors.

```
# Syntax of Nested If-Else
condition_one = True
condition_two = True

if condition_one:
    # Runs if condition_one is True
    pass
    if condition_two:
        # Runs if both are True
        # Logic here
        pass
    else:
        # Runs if condition_one is True but condition_two is False
        # Logic here
        pass
else:
    # Runs if condition_one is False
    # Logic here
    pass
```

**Example 1: Driving License Eligibility**  
**Description:**  
This program verifies if a person can apply for a heavy-vehicle license. It first checks the Age (Outer If). If the person is 18+, it then checks if they already possess a Learner's Permit (Inner If).

```
age = 22
has_permit = True

if age >= 18:
    # This block only runs if age is 18 or above
    if has_permit == True:
        print("Status: Eligible for Permanent License.")
    else:
        print("Status: Please apply for a Learner's Permit first.")
else:
    print("Status: You must be 18 to apply.")
```

**Example 3: Blood Donor Verification**  
**Description:**  
This program evaluates if a person is fit to donate blood. It first checks the Age (Outer If). If the person is 18 or older, it then verifies if their Weight meets the minimum requirement (Inner If).

```
# Taking input for age and weight
age = 20
weight = 55

if age >= 18:
    # This block only runs if age is 18 or above
    if weight >= 50:
        print("Result: You are eligible to donate blood.")
    else:
        print("Result: Sorry, you are underweight for donation.")
else:
    print("Result: You must be at least 18 years old.")
```

**Example 4: ATM Withdrawal Logic**  
**Description:**  
This program simulates how an ATM handles a transaction. It first verifies the PIN entered by the user (Outer If). If the PIN is correct, it then checks if the account balance is sufficient (Inner If) to complete the withdrawal.

```
# ATM Logic Verification
correct_pin = "1122"
balance = 5000

entered_pin = input("Enter your 4-digit PIN: ")

if entered_pin == correct_pin:
    # This block only runs if the PIN is correct
    amount = int(input("Enter amount to withdraw: "))

    if amount <= balance:
        print("Transaction Success: Please collect your cash.")
    else:
        print("Error: Insufficient balance.")
else:
    print("Error: Incorrect PIN. Access denied.")
```

**Example 5: Exam Result with Distinction**  
**Description:**  
This program evaluates a student's final performance. It first checks if the student has passed by scoring 40 or more (Outer If). If they pass, it then checks if their marks are 90 or above (Inner If) to award a Distinction.

```
# Evaluating Exam Results
marks = int(input("Enter your exam marks: "))

if marks >= 40:
    # This block runs only if the student passed
    if marks >= 90:
        print("Result: Passed with Distinction!")
    else:
        print("Result: Passed.")
else:
    print("Result: Failed. Better luck next time.")
```
