### **Q1.** **Variable**
**What is a variable in Python? Write the syntax to create a variable and give an example**.  
**Answer:**  
A variable in Python is like a container that stores data so we can use it later in a program.   
Variables allow us to store numbers, text, or boolean values and refer to them using a name in the program.

**Syntax:**  
variable_name = value

```
# Example of variables
a = 5
student_name = "Batool Sheikh"

print(a)
print(student_name)
```

**Output:**  
5  
Batool Sheikh

**Explanation:**   
The variable **'a'** stores the number 5 and **'student_name'** stores text.
This shows how variables hold data for later use in Python programs.

### **Q7.** **type() Function**  
**What is the type() function in Python? Write an example and show how to use it.**  
**Answer:**  
The type() function is used to check the data type of any variable in Python.  
It returns the type of the variable, such as int, float, str, or bool.

```
# Example of using type() function
x = 10
y = "Hello"

print(type(x))
print(type(y))
```

**Output:**

<class 'int'>  
<class 'str'>  
**Explanation:**   
The first print shows **<class 'int'>** because x is an integer.
The second print shows **<class 'str'>** because y is a string.  
The type() function tells us the data type of a variable.

### **Q9.** **Type Casting**
**What is type casting in Python? Give a Python example to convert a string into an integer.**  
**Answer:**  
Type casting is the process of converting a variable from one data type to another,
such as from string to integer, integer to float, etc. It allows you to change the type of a variable
so it can be used in operations that require a specific data type.

```
# Example of type casting
age_str = "20"         # age as string
age_int = int(age_str)  # convert string to integer
print(age_int)
```

**Output:**  
20  
**Explanation:**  
The variable **'age_str'** is a string containing "20".  
Using int(), it is converted to an integer and stored in 'age_int'.
This demonstrates how type casting works in Python.

### **Q10.** **Arithmetic Operations**
**What are arithmetic operations in Python? Give an example using addition, subtraction, multiplication, and division.**  
**Answer:**  
Arithmetic operations are used to perform mathematical calculations on numeric data.
Python supports addition (+), subtraction (-), multiplication (*), division (/), and more.
These operations can be done on integers and floats.

```
# Example of arithmetic operations
a = 10
b = 5

sum_result = a + b
diff_result = a - b
prod_result = a * b
div_result = a / b

print("Sum:", sum_result)
print("Difference:", diff_result)
print("Product:", prod_result)
print("Division:", div_result)
```

**Output:**  
Sum: 15  
Difference: 5  
Product: 50  
Division: 2.0  
**Explanation:**
- Sum (a + b) adds 10 and 5 to get 15.  
- Difference (a - b) subtracts 5 from 10 to get 5.  
- Product (a * b) multiplies 10 and 5 to get 50.  
- Division (a / b) divides 10 by 5 to get 2.0.  
These are the basic arithmetic operations in Python.
